```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
#SII
#Valores de A
a6716=2.6e-4
a6731=8.8e-4

#Omega
d32_d52=7.47
s32_d52=4.14
s32_d32=2.76
d52_d32=7.47

#Pesos
g52=6
g32=4

T=10000
C=8.62e-6/np.sqrt(T)
ne=np.linspace(1,100000,100000)
```

```python
n=(a6716*g52)*(1+(d32_d52/s32_d52)+(d32_d52/s32_d32)+((a6731*g32)/(C*ne*s32_d32)))

d=(a6731*g32)*(1+(d52_d32/s32_d32)+(d52_d32/s32_d52)+((a6716*g52)/(C*ne*s32_d52)))

j=n/d
```

```python
#OII
#Valores de A
a3729=3.6e-5
a3726=1.6e-4

#Omegas
o_d32_d52=1.17
o_s32_d52=0.804
o_s32_d32=0.536
o_d52_d32=1.17
```

```python
n_o=(a3729*g52)*(1+(o_d32_d52/o_s32_d52)+(o_d32_d52/o_s32_d32)+((a3726*g32)/(C*ne*o_s32_d32)))

d_o=(a3726*g32)*(1+(o_d52_d32/o_s32_d32)+(o_d52_d32/o_s32_d52)+((a3729*g52)/(C*ne*o_s32_d52)))

j_o=n_o/d_o

plt.plot(np.log10(ne),j_o,label='[O II]')
plt.plot(np.log10(ne),j,label = '[S II]')

plt.xlabel('Electron density ($cm^{-3}$)')
plt.ylabel('Intensity ratio')
plt.legend()
plt.show()
```
